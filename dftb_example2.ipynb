```python
import argparse
import os
import shutil
import subprocess
import sys
import textwrap
import json
from pathlib import Path
from ase.io import read, write
from ase.symbols import Symbols
import numpy as np
import matplotlib.pyplot as plt
from jarvis.io.vasp.inputs import Poscar

class DFTBPlusCalculator:
    def __init__(self, dftb_executable, sk_dir, k_mesh, optimise=True,
                 calculate_bandgap=False, calc_dos=True, write_hs=False):
        self.dftb_executable = dftb_executable
        self.sk_dir = sk_dir
        self.k_mesh = k_mesh
        self.optimise = optimise
        self.calculate_bandgap = calculate_bandgap
        self.calc_dos = calc_dos
        self.write_hs = write_hs

    def run(self, cmd, cwd=None, log="run.log"):
        """Run *cmd* in *cwd*, streaming stdout/err to screen & file."""
        print(f" → running: {' '.join(cmd)}")
        print('dir', os.getcwd())
        with open(Path(cwd, log), "a") as fh:
            proc = subprocess.Popen(cmd, cwd=cwd, stdout=subprocess.PIPE,
                                    stderr=subprocess.STDOUT, text=True)
            for line in proc.stdout:
                sys.stdout.write(line)
                fh.write(line)
        if proc.wait() != 0:
            raise RuntimeError(f"Command {' '.join(cmd)} failed.")

    def gen_maxam_from_atoms(self, atoms):
        """Return correct MaxAngularMomentum block (one letter per element)."""
        lmap = {}
        for sym in sorted(set(atoms.get_chemical_symbols())):
            Z = atoms[atoms.get_chemical_symbols().index(sym)].number
            if   Z == 1:                lmap[sym] = "s"            # H
            elif Z <= 10:               lmap[sym] = "p"            # He–Ne (s- and p-SKFs)
            elif Z <= 56:               lmap[sym] = "d"            # transition/block-d
            else:                       lmap[sym] = "f"            # lanthanides/actinides
        return "\n".join(f"{el} = \"{l}\"" for el, l in lmap.items())

    def supercell_folding_block(self, n1, n2, n3, shift=(0.0, 0.0, 0.0)):
        """Return a valid SuperCellFolding string for dftb_in.hsd."""
        mat = f"{n1} 0 0\n 0 {n2} 0\n 0 0 {n3}"
        sh  = f"{shift[0]} {shift[1]} {shift[2]}"
        return f"{mat}\n {sh}"

    def dict_to_hsd(self, d, indent=0, top_level_key=None):
        """Convert a nested dictionary to HSD format, top-level key can be block (no `=`)."""
        lines = []
        prefix = "  " * indent

        if top_level_key:
            lines.append(f"{top_level_key} {{")
            indent += 1
            prefix = "  " * indent

        for key, value in d.items():
            if isinstance(value, dict):
                lines.append(f"{prefix}{key} = {{")
                lines.extend(self.dict_to_hsd(value, indent + 1))
                lines.append(f"{prefix}}}")
            elif isinstance(value, list):
                lines.append(f"{prefix}{key} = [")
                for item in value:
                    if isinstance(item, dict):
                        lines.append(f"{prefix}  {{")
                        lines.extend(self.dict_to_hsd(item, indent + 2))
                        lines.append(f"{prefix}  }}")
                    else:
                        lines.append(f"{prefix}  {item}")
                lines.append(f"{prefix}]")
            else:
                lines.append(f"{prefix}{key} = {value}")

        if top_level_key:
            lines.append("}")

        return lines

    def write_hsd(self, fname, geometry_ref, maxl_block):
        # Geometry block
        geometry_block = f"""Geometry = GenFormat {{
    <<< {geometry_ref}
}}"""

        # Driver block
        driver_lines = []
        if self.optimise:
            driver_dict = {
                "GeometryOptimization": {
                    "Optimizer": "Rational {}",
                    "MovedAtoms": "1:-1",
                    "MaxSteps": 100,
                    "OutputPrefix": '"geom.out"',
                    "Convergence": {
                        "GradElem": "1E-4"
                    }
                }
            }
            driver_lines = self.dict_to_hsd(driver_dict, top_level_key="Driver")

        # MaxAngularMomentum and KPoints (raw blocks)
        maxl_block_str = f"""  MaxAngularMomentum = {{
{maxl_block}
}}"""

        kfold_block_str = f"""  KPointsAndWeights = SuperCellFolding {{
    {self.supercell_folding_block(*self.k_mesh)}
}}"""

        # Hamiltonian block
        hamiltonian_dict = {
            "SCC": "Yes",
            "Filling": {
                "Fermi": {
                    "Temperature [Kelvin]": 0
                }
            },
            "SlaterKosterFiles": {
                "Type2FileNames": {
                    "Prefix": f'"{self.sk_dir}/"',
                    "Separator": '"-"',
                    "Suffix": '".skf"'
                }
            }
        }

        # Analysis block (always include CalculateForces)
        analysis_dict = {
            "CalculateForces": "Yes"
        }

        if self.calc_dos:
            analysis_dict["ProjectStates"] = {
                "Region": {
                    "Atoms": "1:-1",
                    "OrbitalResolved": "No"
                }
            }

        if self.calculate_bandgap:
            analysis_dict["BandStructure"] = {
                "OutputKpoints": "Yes",
                "BandLines": {
                    "BrillouinZone": {
                        "SpecialPoints": {
                            "Gamma": "[ 0.0 0.0 0.0 ]",
                            "X": "[ 0.5 0.0 0.0 ]",
                            "L": "[ 0.5 0.5 0.5 ]",
                            "W": "[ 0.5 0.25 0.75 ]",
                            "K": "[ 0.375 0.375 0.75 ]"
                        },
                        "Lines": [
                            {"From": "Gamma", "To": "X", "Points": 20},
                            {"From": "X", "To": "L", "Points": 20},
                            {"From": "L", "To": "Gamma", "Points": 20},
                            {"From": "Gamma", "To": "W", "Points": 20},
                            {"From": "W", "To": "K", "Points": 20}
                        ]
                    }
                }
            }

        analysis_lines = self.dict_to_hsd(analysis_dict, top_level_key="Analysis")

        # Options block
        options_dict = {
            "Options": {
                "WriteResultsTag": "Yes"
            }
        }
        if self.write_hs:
            options_dict["Options"]["WriteHS"] = "Yes"

        # ParserOptions block
        parser_options = {
            "ParserOptions": {
                "ParserVersion": 12
            }
        }

        # Final assembly
        hsd_lines = []
        hsd_lines.append(geometry_block)
        hsd_lines.extend(driver_lines)

        hsd_lines.append("Hamiltonian = DFTB {")
        hsd_lines.extend(self.dict_to_hsd(hamiltonian_dict, indent=1))
        hsd_lines.append(maxl_block_str)
        hsd_lines.append(kfold_block_str)
        hsd_lines.append("}")

        hsd_lines.extend(analysis_lines)
        hsd_lines.extend(self.dict_to_hsd(options_dict))
        hsd_lines.extend(self.dict_to_hsd(parser_options))

        with open(fname, "w") as f:
            f.write("\n".join(hsd_lines) + "\n")

    def get_energy(self, path="detailed.out"):
        with open(path) as f:
            for line in f:
                if "Total energy:" in line:
                    return float(line.split()[-2]) * 27.2114

    def get_forces(self, path="detailed.out"):
        forces = []
        with open(path) as f:
            lines = f.readlines()
            for i, line in enumerate(lines):
                if "Total Forces" in line:
                    for l in lines[i+2:]:
                        if not l.strip():
                            break
                        parts = l.split()
                        forces.append([float(x) * 51.422086 for x in parts[-3:]])
                    break
        return np.array(forces)

    def read_band_out(self, file_path):
        """Read band structure data from band.out file."""
        with open(file_path, 'r') as f:
            lines = f.readlines()

        data = []
        for line in lines:
            if line.strip().startswith('KPT') or line.strip() == '':
                continue
            try:
                data.append([float(x) for x in line.split()])
            except ValueError:
                break

        data = np.array(data)
        k_points = data[:, 0]
        energies = data[:, 1:]
        return k_points, energies

    def gaussian(self, x, mu, sigma):
        """Gaussian function for broadening."""
        return np.exp(-((x - mu) / sigma)**2 / 2) / (sigma * np.sqrt(2 * np.pi))

    def calculate_dos(self, energies, sigma, energy_range, num_points):
        """Calculate density of states using Gaussian broadening."""
        dos = np.zeros(num_points)
        energies_flat = energies.flatten()
        energy_grid = np.linspace(energy_range[0], energy_range[1], num_points)
        for energy in energies_flat:
            dos += self.gaussian(energy_grid, energy, sigma)
        return energy_grid, dos

    def get_bandgap(self, path="band.out"):
        k_points, energies = self.read_band_out(path)
        # Simple bandgap calculation, might need to be adjusted
        vbmax = np.max(energies[:, :energies.shape[1]//2])
        cbmin = np.min(energies[:, energies.shape[1]//2:])
        return cbmin - vbmax, cbmin, vbmax

    def run_dftb(self, atoms, optdir):
        if not os.path.exists(optdir):
            os.mkdir(optdir)

        cwd = os.getcwd()
        my_cwd = os.getcwd()
        os.chdir(cwd)
        write(optdir + "/geo.gen", atoms, format="gen")
        maxl_block = self.gen_maxam_from_atoms(atoms)
        self.write_hsd(optdir + "/dftb_in.hsd", "geo.gen", maxl_block)
        self.run([self.dftb_executable], cwd=optdir)
        os.chdir(my_cwd)

        # Postprocessing
        final_atoms = read(optdir + "/geom.out.gen", format="gen")
        energy = self.get_energy(path=optdir + "/detailed.out")
        forces = self.get_forces(path=optdir + "/detailed.out")

        # Calculate DOS
        if self.calc_dos:
            k_points, energies = self.read_band_out(optdir + "/band.out")
            sigma = 0.1  # eV, broadening parameter
            energy_range = (-5, 5)  # eV, energy range for DOS
            num_points = 1000  # number of points in the DOS energy grid
            energy_grid, dos = self.calculate_dos(energies, sigma, energy_range, num_points)

            # Calculate bandgap
            bandgap, cbm, vbm = self.get_bandgap(path=optdir + "/band.out")

            # Save results to out.json
            results = {
                "energy": energy,
                "forces": forces.tolist(),
                "dos": {"energy_grid": energy_grid.tolist(), "dos": dos.tolist()},
                "bandgap": bandgap,
                "cbm": cbm,
                "vbm": vbm
            }
            with open(optdir + "/out.json", "w") as f:
                json.dump(results, f, indent=4)

        return final_atoms, energy, forces

def main():
    poscar_string = """bulk@mp-149
1.0
3.364155 -0.000000 1.942295
1.121385 3.171756 1.942295
-0.000000 -0.000000 3.884592
Si
2
direct
0.875000 0.875000 0.875000 Si
0.125000 0.125000 0.125000 Si
    """
    #poscar_path = 'POSCAR'
    atoms = Poscar.from_string(poscar_string).atoms
    atoms = atoms.ase_converter()

    dftb_executable = "/usr/local/bin/dftb+"
    sk_dir = os.path.abspath("/content/siband-1.1.0/skfiles/")
    k_mesh = [20, 20, 20]
    optdir = "1_opt"

    calculator = DFTBPlusCalculator(dftb_executable, sk_dir, k_mesh)
    final_atoms, energy, forces = calculator.run_dftb(atoms, optdir)
    print("Energy (eV):", energy)
    print("Forces (eV/Å):", forces)

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import re
%matplotlib inline


#####extract the fermi level in eV####
fermi_file_path = "./1_opt/detailed.out"

# Read file content
with open(fermi_file_path, 'r') as file:
    lines = file.readlines()

# Extract numeric value in eV using regex
fermi_eV = None
for line in lines:
    if 'Fermi' in line and 'eV' in line:
        match = re.search(r'([-+]?\d*\.\d+|\d+)\s*eV', line)
        if match:
            fermi_eV = float(match.group(1))
        break

####Preprocessing####
# Load the input file and remove "|" characters
file_path = './1_opt/dos_total.dat'

# Data storage
x_values = []
y_values = []

# Function to check if a line starts with a valid number
def starts_with_number(token):
    try:
        float(token)
        return True
    except ValueError:
        return False

# Load unique x and y values
with open(file_path, 'r') as file:
    for line in file:
        tokens = line.split()
        if tokens and starts_with_number(tokens[0]):  # Check if the first token is a valid number
            x = float(tokens[0])
            x_values.append(x)

# Load all y values
with open(file_path, 'r') as file:
    for line in file:
        tokens = line.split()
        if tokens and starts_with_number(tokens[0]):  # Check if the first token is a valid number
            y = float(tokens[1])
            y_values.append(y)

# Convert x and y values to numpy arrays
npx = np.array(x_values)-fermi_eV
npy = np.array(y_values)

# Check if reshaping is possible
num_bands = len(y_values) // len(x_values)
if len(y_values) % len(x_values) != 0:
    raise ValueError("Mismatch between the number of y-values and x-values. Check your data.")

# Reshape y-values into a 2D array: rows = x_values, columns = bands
ref_npy = npy.reshape(num_bands, len(x_values))

# Plotting
#plt.figure(figsize=(10, 6))
plt.clf()
for i in range(ref_npy.shape[0]):  # Iterate over rows (bands)
    plt.plot(npx, ref_npy[i, :], color='k')
plt.xlabel('Energy(E-Ef)')
plt.ylabel('DOS(arb. unit)')
plt.xlim(-4,4)
plt.legend()
plt.rcParams.update({'font.size': 22})
#plt.show()
plt.savefig('dftb_dosfromrelax.png',bbox_inches='tight')
```
