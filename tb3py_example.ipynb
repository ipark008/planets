```python
import os
import re
import spglib
import numpy as np
import matplotlib.pyplot as plt
from jarvis.core.kpoints import generate_kgrid
from jarvis.io.vasp.inputs import Poscar
from jarvis.core.kpoints import Kpoints3D
from jarvis.io.wannier.outputs import WannierHam
from jarvis.io.vasp.outputs import Vasprun
from jarvis.db.figshare import data
from jarvis.core.atoms import Atoms
from matplotlib.gridspec import GridSpec


####write input file for julia and run the script####
lines="""using ThreeBodyTB
c = makecrys("POSCAR")
###Relaxation######
crys_new, tbc_updated, energy, force, stress = relax_structure(c);
####SCF calculation#####
energy, tbc, flag = scf_energy(crys_new,smearing=0.002);
println("directgap, indirectgap, gaptype, bandwidth",string(ThreeBodyTB.BandStruct.band_summary(tbc)))
println(tbc)
write_hr_dat(tbc,filename="my_hr.dat")
####DOS calculation#####
ret =dos(tbc,smearing=0.002, grid=missing, do_display=false) #default 0.04
using DelimitedFiles
column2 = vec(ret[2])
data = [ret[1] column2]
writedlm("dos_tb3py.txt", data, '\\t')
"""
with open("tb3py_in.jl","w") as f:
     f.write(lines)
```

```python
###extract the fermi level in eV###
fermi_file_path = "./tb3py.out"

# Read file content
with open(fermi_file_path, 'r') as file:
    lines = file.readlines()

# Extract numeric value in eV using regex
fermi_tb = None
for line in lines:
    if 'efermi' in line and 'eV' in line:
        match = re.search(r'([-+]?\d*\.\d+|\d+)\s*eV', line)
        if match:
            fermi_tb = float(match.group(1))
        break


#####################Extract band-data from TB3PY#####################
w = WannierHam(filename='my_hr.dat')

with open('POSCAR', 'r') as f:
    pos = f.read()

atoms=Poscar.from_string(pos).atoms
eigs = []
kpoints, labels = Kpoints3D().interpolated_points(atoms)
for i in kpoints:
    # print (i)
    val, vect, p = w.solve_ham(k=i, proj=None)
    eigs.append(val)
eigs_tb=np.array(eigs)-fermi_tb

#####################extract band-data from VASP######################
vrun = Vasprun('vasprun.xml')
spin = 0  # Assuming spin is predefined
fermi_dft=vrun.efermi
kpoints = vrun.kpoints.kpts  # Extract k-points for clarity
eigs = np.array(vrun.eigenvalues[spin][0, :, 0])  # First k-point's eigenvalues
# Iterate through k-points and eigenvalues, starting from the second k-point
for kp, j in enumerate(vrun.eigenvalues[spin][1:], start=1):
    eigs = np.column_stack((eigs, j[:, 0]))  # Stack eigenvalues for each k-point
eigs_dft=eigs-fermi_dft


##################Plotting bands and error distribution####################
info = vrun.get_bandstructure(kpoints_file_path='KPOINTS',plot=True)
kpts=info['kp_labels_points']
label=info['kp_labels']

plt.clf()
the_grid=GridSpec(1,2)
plt.rcParams.update({'font.size':14})
plt.figure(figsize=(10,4))

plt.subplot(the_grid[0,0])
plt.title('(c)')
plt.ylabel('Energy(eV)')
plot=True
min_arr = []
dd={}
energy_tol=4
kpoints=eigs_dft.shape[1]
erange = [-energy_tol, energy_tol]
for k in range((kpoints)-1):
            for n in eigs_dft.T[k]:
                diff_arr = []
                if n > erange[0] and n < erange[1]:
                    for v in eigs_tb[k]:
                        diff = abs(n - v)
                        diff_arr.append(diff)
                if diff_arr != []:
                    tmp = np.min(diff_arr)
                    dd.setdefault(n,tmp)
                    min_arr.append(tmp)
maxdiff = "na"
if min_arr != []:
            # print ('min_arr',min_arr)
            print("MAX diff", max(min_arr))
            maxdiff = max(min_arr)
print("maxdiff", maxdiff)
if plot:
            for i, ii in enumerate(eigs_tb.T):
                plt.plot(ii, color="b")
            for i, ii in enumerate(eigs_dft):
                plt.plot(ii, color="r")
            plt.ylim([-energy_tol, energy_tol])
            #if kp_labels_points != [] and kp_labels != []:
            #    plt.xticks(kp_labels_points, kp_labels)
#plt.show()
plt.xlim(kpts[0],kpts[-1])
plt.xticks(kpts, label)
plt.subplot(the_grid[0,1])
plt.title('(d)')
plt.plot(list(dd.values()),list(dd.keys()),'.')
plt.ylabel('Energy(eV)')
plt.xlabel('TB3PY-DFT ($\delta$) (eV)')
plt.savefig('comparison.png',bbox_inches='tight')
plt.show()

###########################DOS extraction and plotting########################
####Preprocessing####
# Load the input file and remove "|" characters
file_path = './dos_tb3py.txt'

# Data storage
x_values = []
y_values = []

# Function to check if a line starts with a valid number
def starts_with_number(token):
    try:
        float(token)
        return True
    except ValueError:
        return False

# Load unique x and y values
with open(file_path, 'r') as file:
    for line in file:
        tokens = line.split()
        if tokens and starts_with_number(tokens[0]):  # Check if the first token is a valid number
            x = float(tokens[0])
            x_values.append(x)

# Load all y values
with open(file_path, 'r') as file:
    for line in file:
        tokens = line.split()
        if tokens and starts_with_number(tokens[0]):  # Check if the first token is a valid number
            y = float(tokens[1])
            y_values.append(y)

# Convert x and y values to numpy arrays
npx = np.array(x_values)#-fermi_tb
npy = np.array(y_values)

# Check if reshaping is possible
num_bands = len(y_values) // len(x_values)
if len(y_values) % len(x_values) != 0:
    raise ValueError("Mismatch between the number of y-values and x-values. Check your data.")

# Reshape y-values into a 2D array: rows = x_values, columns = bands
dos = npy.reshape(num_bands, len(x_values))
energy = npx

# Plotting
#plt.figure(figsize=(10, 6))
plt.clf()
for i in range(dos.shape[0]):  # Iterate over rows (bands)
    plt.plot(energy, dos[i, :], color='k')
plt.xlabel('Energy(E-Ef)')
plt.ylabel('DOS(arb. unit)')
plt.xlim(-8,8)
plt.legend()
plt.rcParams.update({'font.size': 22})
#plt.show()
plt.savefig('tb3py_dos.png',bbox_inches='tight')
```
